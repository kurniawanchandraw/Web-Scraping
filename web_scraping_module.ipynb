### <a id='toc1_1_1_'></a>[**1.1. Berkenalan dengan HTML**](#toc0_)

**Hypertext Markup Language (HTML)** adalah bahasa *markup* standar untuk dokumen yang dirancang agar dapat ditampilkan di peramban web. HTML mendeskripsikan struktur sebuah halaman web dan dapat digunakan bersama dengan **Cascading Style Sheets (CSS)** serta bahasa pemrograman seperti **JavaScript** untuk membuat situs web interaktif. HTML terdiri dari serangkaian elemen yang "memberi tahu" peramban bagaimana cara menampilkan kontan. Teakhir, elemen-elemen dalam HTML direpresentasikan dengan *tag*.

HTML, CSS, dan JavaScript adalah tiga bahasa yang berbeda tetapi digunakan bersama untuk membuat halaman web. Setiap halaman web di internet dibangun dengan **kombinasi** dari ketiga bahasa ini. Masing-masing dari ketiga bahasa pemrograman ini memiliki peran yang berbeda dalam pembuatan halaman web. Gambar di bawah ini menunjukkan perbedaan fungsi antara HTML, CSS, dan JavaScript.

HTML adalah tulang punggung dari konten web. Ini adalah bahasa *markup* yang digunakan untuk **menyusun dan mendefinisikan elemen-elemen dalam sebuah halaman web**, seperti *heading*, paragraf, gambar, tautan, daftar, dan lainnya. HTML menggunakan *tag* (dikelilingi oleh tanda kurung sudut) untuk menunjukkan bagaimana elemen-elemen harus ditampilkan di halaman web. Misalnya, `\<h1>` menunjukkan *heading* tingkat atas, `\<p>` merepresentasikan sebuah paragraf, dan `\<img>` digunakan untuk menampilkan gambar. HTML menyediakan struktur dan konten sebuah halaman web, berfungsi sebagai dasar dalam pengembangan web.

<div style="border-left: 5px solid #2980b9; padding: 10px; background-color: #d6eaf8;">
<b> Apa Itu Bahasa <i>Markup</i>?</b>

Bahasa <i>markup</i> adalah jenis bahasa yang digunakan untuk menandai atau mencetak elemen dalam sebuah dokumen agar dapat ditampilkan dengan format tertentu. Tidak seperti bahasa pemrograman yang berisi logika dan perhitungan, bahasa markup hanya berfungsi untuk menyusun struktur dan tampilan sebuah dokumen, terutama pada web. Contohnya adalah HTML, XML, Markdown, dan $\LaTeX$.
</div>


Berikut adalah beberapa tag dalam HTML:  
* `<!DOCTYPE html>` deklarasi ini mendefinisikan bahwa dokumen ini adalah HTML5.  
* `<html>` elemen ini adalah elemen akar dari sebuah halaman HTML.  
* `<div>` tag ini mendefinisikan sebuah divisi atau bagian dalam dokumen HTML. Biasanya digunakan sebagai wadah untuk elemen lainnya.  
* `<head>` elemen ini berisi informasi *meta* tentang dokumen.  
* `<title>` elemen ini menentukan judul untuk dokumen.  
* `<body>` elemen ini berisi konten halaman yang dapat dilihat oleh pengguna.  
* `<h1>` elemen ini mendefinisikan heading atau judul besar.  
* `<p>` elemen ini mendefinisikan sebuah paragraf.  
* `<a>` elemen ini mendefinisikan sebuah *hyperlink*.  

*Tag* HTML biasanya terdiri dari **pasangan**, seperti `<p>` dan `</p>`. *Tag* pertama dalam pasangan disebut ***tag* pembuka**, sedangkan *tag* kedua disebut ***tag* penutup**. *Tag* penutup ditulis seperti tag pembuka, tetapi dengan tambahan garis miring (`/`) sebelum nama tag.

## <a id='toc1_2_'></a>[**2. Pengantar *Web Scraping***](#toc0_)

<div style="border-left: 5px solid #2980b9; padding: 10px; background-color: #d6eaf8;">
<b> Apa Itu <i>Web Scraping</i>?</b>

*Web scraping* adalah proses mengumpulkan informasi dari internet. Bahkan menyalin dan menempelkan lirik lagu favorit pun dapat dianggap sebagai bentuk sederhana dari *web scraping*! Namun, istilah "web scraping" biasanya merujuk pada proses yang melibatkan otomatisasi. 

Meskipun beberapa situs web tidak memperbolehkan *scraper* otomatis mengambil data mereka—yang bisa menimbulkan masalah hukum—ada juga yang tidak mempermasalahkannya.
</div>


Internet telah berkembang secara organik dari berbagai sumber. Internet menggabungkan berbagai teknologi, gaya, dan kepribadian yang berbeda, serta terus berkembang setiap hari. Dengan kata lain, internet sangat luas. Oleh karena itu, akan terdapat beberapa tantangan saat melakukan *web scraping*:

1. **Keberagaman**: Setiap situs web berbeda. Meskipun ada pola umum yang sering muncul, setiap situs memiliki struktur uniknya sendiri dan memerlukan pendekatan khusus jika ingin mengekstrak informasi yang relevan.

2. **Ketahanan**: Situs web terus berubah.

![Ilustrasi Web Scraping](https://miro.medium.com/v2/resize:fit:1024/0*pGYcGfBaRcM4i0H5.jpg)

Dalam *web scraping*, *scraper* bekerja dengan cara:
- **Mengakses halaman web**; mengirim permintaan HTTP ke situs web target.
- **Mengambil HTML-nya**; mengunduh kode sumber HTML dari halaman tersebut.
- **Memilah informasi**; menggunakan pustaka seperti `BeautifulSoup` atau `lxml` untuk mengekstrak data yang diperlukan.
- **Menyimpan hasilnya**; data yang diekstrak bisa disimpan dalam CSV, *database*, atau file lain untuk analisis lebih lanjut.

### <a id='toc1_2_1_'></a>[**2.1. Alternatif *Web Scraping*: API**](#toc0_)

<div style="border-left: 5px solid #2980b9; padding: 10px; background-color: #d6eaf8;">
<b> Apa Itu API?</b>

API adalah singkatan dari **Application Programming Interface**. Secara sederhana, API berfungsi sebagai **lapisan komunikasi** atau **antarmuka** yang memungkinkan berbagai sistem untuk saling berkomunikasi tanpa harus memahami secara detail bagaimana masing-masing sistem bekerja. API dapat disebut juga sebagai kumpulan fungsi yang dapat **dipanggil atau dijalankan** oleh program lain.

> Antarmuka: bagian yang sama antara dua atau lebih komponen terpisah pada sebuah sistem komputer.
</div>


API biasanya digunakan untuk membagi (*share*) data antara dua atau lebih sistem yang **tidak berhubungan**.

![Ilustrasi API](https://voyager.postman.com/illustration/diagram-what-is-an-api-postman-illustration.svg)

Arsitektur API biasanya dijelaskan dalam kaitannya dengan klien dan server. Aplikasi yang mengirimkan permintaan disebut sebagai klien dan aplikasi yang mengirimkan respons disebut sebagai server. Sebagai contoh, basis data cuaca BMKG adalah servernya sedangkan aplikasi seluler adalah kliennya.

API bekerja dengan cara menerima permintaan (`request`) dari suatu aplikasi, lalu mengembalikan respons (`response`) yang berisi data atau hasil yang diminta. Biasanya, komunikasi ini terjadi melalui internet menggunakan protokol seperti HTTP atau HTTPS.

</div>


Web Service adalah jenis API yang berjalan di internet menggunakan protokol web. HTTP dan HTTPS adalah protokol komunikasi yang digunakan untuk mentransfer data antara browser (*client*) dan server (*backend*) dalam jaringan internet.

![Ilustrasi Web Service](https://dicoding-assets.sgp1.cdn.digitaloceanspaces.com/blog/wp-content/uploads/2020/05/web-service.jpg)

### <a id='toc1_2_2_'></a>[**2.2. Etika dalam *Web Scraping***](#toc0_)

*Web scraping* adalah alat yang sangat kuat dan harus digunakan secara bertanggung jawab. Di bawah ini adalah beberapa hal yang perlu dipertimbangkan sebelum kita mulai melakukan *scraping data*:  

- **Selalu periksa file `robots.txt` dari situs yang akan kita *scrape***, karena file ini berisi panduan tentang bagaimana bot seharusnya berperilaku di situs tersebut.  
- **Jangan mengirimkan terlalu banyak permintaan dalam waktu singkat ke *website***, karena hal itu dapat membebani server mereka dan/atau dikategorikan sebagai **serangan DDoS**. Jika kita perlu melakukan *scraping* pada banyak halaman, kita bisa **membatasi kecepatan permintaan secara buatan**, seperti yang akan kami tunjukkan dalam kode.  
- **Jangan melakukan pembajakan (*piracy*) atau penggunaan komersial yang tidak sah terhadap data yang kita ekstrak.**  
- **Selain itu, beberapa perusahaan dan *website* mungkin sudah menyediakan data yang kita butuhkan dalam bentuk API resmi.** Jika ada **API publik yang dapat menyediakan informasi yang kita cari**, maka ***web scraping* sebaiknya dihindari** sepenuhnya.  

Lebih lanjut, baca artikel **James Densmore** tentang [Etika dalam *Web Scraping*](https://medium.com/data-science/ethics-in-web-scraping-b96b18136f01).

- [Requests](https://docs.python-requests.org/en/latest/): Pustaka sederhana namun kuat untuk mengirim berbagai jenis permintaan HTTP ke server web.
- [Beautiful Soup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/): Pustaka untuk mem-*parsing* dokumen HTML dan XML. Pustaka ini bekerja dengan *parser* yang dipilih pengguna untuk menyediakan cara yang mudah dalam menavigasi, mencari, dan memodifikasi struktur *parse tree*.

### <a id='toc1_3_1_'></a>[**3.1. Studi Kasus: Web Books to Scrape**](#toc0_)

#### <a id='toc1_3_1_1_'></a>[**Langkah 1: Mendapatkan HTML**](#toc0_)

```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
response = requests.get(url)

if response.status_code == 200:
    print('Request Berhasil!')
else:
    print('Request Gagal! Status Code: ', response.status_code)
```

Catatan:
1. Kode `200` dijadikan sebagai indikator, merujuk pada [panduan kode status HTTP](https://en.wikipedia.org/wiki/List_of_HTTP_status_codes).
2. Beberapa halaman membutuhkan *user agent*. *User agent* dapat diketahui dengan mudah melalui [laman web ini](https://www.whatismybrowser.com/detect/what-is-my-user-agent/#google_vignette). Untuk menambahkan *user agent*, gunakan kode `requests.get(url, headers={'user_agent': user_agent})`. 

Mengecek tipe dari variabel `response`:

```python
type(response)
```

#### <a id='toc1_3_1_2_'></a>[**Langkah 2: Parsing HTML**](#toc0_)

Setelah requests berhasil, kita dapat mengonversi objek `response` menjadi objek `BeautifulSoup`, yang merepresentasikan dokumen dalam bentuk struktur data bertingkat. Kita membuat objek `BeautifulSoup` dengan memberikan dua argumen:
- Konten HTML mentah dalam bentuk byte (menggunakan atribut `.content`).
- *Parser* HTML yang ingin kita gunakan.

Terdapat tiga pilihan *parser* yang dapat digunakan dengan `BeautifulSoup`: `html.parser`, `lxml`, dan `html5lib`.

```python
html = response.content
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
soup
```

Saran yang baik untuk melihat *parse tree* adalah menggunakan metode `prettify()`, yang mengubah *parse tree* dari `BeautifulSoup` menjadi string Unicode yang diformat dengan rapi, dengan setiap *tag* dan teks ditempatkan pada baris terpisah.  Sebab hasilnya bisa sangat panjang, kita hanya akan mencetak 1.500 karakter pertama.

```python
print(soup.prettify()[:1500])
```

Apabila tidak terdapat perbedaan, kode HTML situs sudah menggunakan indentasi yang baik dan struktur yang bersih.

#### <a id='toc1_3_1_3_'></a>[**Langkah 3: Mengekstrak Data**](#toc0_)

Dalam `bs4`, mengekstrak data dapat dilakukan dengan fungsi `find`. Kode berikut untuk memahami perbedaan antara `find` dengan `find_all`:

```python
test1 = """
<ul>
<li>First</li>
<li>Second</li>
<li>Third</li>
</ul>
"""   
test2 = BeautifulSoup(test1,'html.parser')

for n in test2.find('li'):
  # It Give you one element     
  print(n)

for n in test2.find_all('li'):    
  # It Give you all elements
  print(n)
```

Perhatikan perbedaan tipe variabel `find` dan `find_all` berikut.

```python
type(test2.find('li'))
```

```python
type(test2.find_all('li'))
```

#### <a id='toc1_3_1_4_'></a>[**Langkah 4: Menyimpan Semua Data**](#toc0_)

Perhatikan bahwa data yang disimpan pada langkah sebelumnya adalah data pada halaman pertama dan buku pertama saja. Untuk menyimpan semua data, dapat memanfaatkan `for-loop`.

```python
import pandas as pd

# 01. Buat dataframe kosong
df = pd.DataFrame(columns=['Title', 'Price', 'Stock', 'Rating'])

# 02. Inisiasi loop untuk mengambil data pada setiap halaman
total_page = 50
all_books = []

for page in range(1, total_page+1):
    # 03. Mengakses URL setiap halaman
    url = f'https://books.toscrape.com/catalogue/page-{page}.html'

    # 04. Request ke URL
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # 05. Mengambil data pada setiap artikel
    for article in soup.find_all('article', {'class': 'product_pod'}):
        title = article.find('h3').find('a').get('title')
        price = article.find('p', {'class': 'price_color'}).get_text()
        stock = article.find('p', {'class': 'instock availability'}).get_text().strip()
        rating = article.find('p', {'class': 'star-rating'}).get('class')[1]

        # 06. Menambahkan data ke dataframe
        all_books.append([title, price, stock, rating])

df = pd.DataFrame(all_books, columns=['Title', 'Price', 'Stock', 'Rating'])
```

## <a id='toc1_4_'></a>[**4. API *Web Scraping***](#toc0_)

API Scraping adalah teknik mengambil data dari sebuah situs web melalui API, bukan melalui HTML *parsing*.

| **Metode**                | **Cara Kerja** | **Kelebihan** | **Kekurangan** |
|--------------------------|--------------|--------------|--------------|
| **Web Scraping (HTML *Parsing*)** | Mengambil data dari tampilan HTML dengan BeautifulSoup/Selenium. | Bisa *scraping* semua *website*. | Bisa diblokir, lambat, dan butuh parsing HTML. |
| **API Scraping**         | Mengambil data langsung dari API yang disediakan *website* (format JSON/XML). | Cepat, bersih, legal (jika API resmi). | Tidak semua *website* menyediakan API. |

Terdapat beberapa cara dalam mengakses API, seperti dengan `cURL` ataupun `requests`. `cURL` digunakan untuk mengirim permintaan HTTP ke API dan mendapatkan respons dari server.

> Untuk mengakses API dalam proses *scraping*, kita harus tahu alamat *endpoint* API terlebih dahulu.

*Endpoint* API adalah URL yang digunakan untuk mengakses atau berinteraksi dengan API, berfungsi sebagai lokasi digital spesifik di mana permintaan dikirim dan respons diterima.

![Ilustrasi API Endpoint](https://www.akamai.com/site/en/images/article/2024/what-is-an-api-endpoint.png)

*Endpoint* API berfungsi sebagai gerbang untuk komunikasi antar aplikasi, yang menyediakan lokasi data atau sumber daya yang tepat. Pada dasarnya, *endpoint* API adalah URL khusus yang memungkinkan aplikasi atau sistem mengakses sumber daya atau layanan yang ditawarkan oleh aplikasi atau sistem lain. URL *endpoint* ini memungkinkan aplikasi klien membuat permintaan HTTP seperti GET, POST, PUT, atau DELETE saat berinteraksi dengan API.

Alamat API (*endpoint*) berada di server web, dan kita bisa menemukannya dengan cara tertentu, seperti melihat **dokumentasi API** atau menggunakan **Developer Tools** di browser. Contoh-contoh dari dokumentasi API dapat dilihat pada [Daftar API Lokal Indonesia](https://github.com/farizdotid/DAFTAR-API-LOKAL-INDONESIA).
